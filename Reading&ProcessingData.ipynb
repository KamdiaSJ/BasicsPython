```python
# #PIVOT TABLE
# #in pandas we can create a pivot table just as easily in excel
# the index means the rows, columns are things that you want to see at the top of dataframe
# values are the values and the aggfunc is how you want to aggregate

df.pivot_table?
```

```python
df.pivot_table(
    index='year_py',
    columns='origin',
    values='mpg',
    aggfunc='mean')
```

```python
#to create a visual plot line for the same you can just add .plot.line()
# NOTE: plot function would always treat index as x
df.pivot_table(
    index='year_py',
    columns='origin',
    values='mpg',
    aggfunc='mean').plot.line()
```
